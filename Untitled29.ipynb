```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
def load_data(data_path):
    """
    Load data from a file.

    Parameters:
    data_path (str): Path to the data file.

    Returns:
    pd.DataFrame: Loaded data as a pandas DataFrame.

    Raises:
    ValueError: If the file format is not supported (only CSV and Excel).
    """
    if data_path.endswith('.csv'):
        print(f"CSV Data Was Loaded from {data_path}")
        return pd.read_csv(data_path)
    
    elif data_path.endswith('.xlsx'):
        print(f"Excel Data Was Loaded from {data_path}")
        return pd.read_excel(data_path)
    
    else:
        raise ValueError("Unsupported file format. Only CSV and Excel files are supported.")
```

```python
# Prompt the user for the data file path
#data_path = input("Please provide the path to your data file (CSV or Excel): ")
data_path="International_Report_Departures.csv"
# Load data 
try:
    df = load_data(data_path)
except Exception as e:
    print("An error occurred while loading:", str(e))
    exit(1)    
```

```python
def preprocess_data(df):
    """
    Preprocess the given DataFrame.

    Parameters:
    df (pd.DataFrame): Input DataFrame.

    Returns:
    pd.DataFrame: Preprocessed DataFrame.
    """
    # Impute missing 'carrier' values with the most frequent value
    imputer = SimpleImputer(strategy='most_frequent')
    df['carrier'] = imputer.fit_transform(df[['carrier']])
    
    # Drop rows with missing values
    df.dropna(axis=0, inplace=True)
    
    # Sort the DataFrame by 'data_dte' in ascending order
    df = df.sort_values('data_dte', ascending=True).reset_index(drop=True)
    
    # Reset the index and adjust it to start from 1
    df.index = df.index + 1
    
    # Map 'carriergroup' values from 1 to 'domestic' and 0 to 'foreign'
    df['carriergroup'] = df['carriergroup'].replace({1: 'domestic', 0: 'foreign'})
    
    # Filter out data for the year 2020
    df = df[df['Year'] != 2020]
    print("Data Preprocessing Complete Well Done !")
    return df
```

```python
# Apply data preprocessing
try:
    df = preprocess_data(df)
except Exception as e:
    print("An error occurred during data preprocessing:", str(e))
    exit(1)
```

```python
def visualize_total_flights_by_year(df):
    """
    Visualize total flights by year using a bar plot.

    Parameters:
    df (pd.DataFrame): DataFrame containing flight data.

    Returns:
    None
    """
    df_year = df.groupby('Year')['Total'].sum().reset_index()
    
    sns.set(style='whitegrid')
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Year', y='Total', data=df_year, color='deepskyblue')
    
    plt.ticklabel_format(style='plain', axis='y')
    plt.xlabel('Year')
    plt.ylabel('Total Flights')
    plt.title('Total Flights by Year')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Visualize total flights by year
try:
    visualize_total_flights_by_year(df)
except Exception as e:
    print("An error occurred while visualizing total flights by year:", str(e))
```

```python
def visualize_flights_per_season(df):
    """
    Visualize flights per season across years using bar plots.

    Parameters:
    df (pd.DataFrame): DataFrame containing flight data.

    Returns:
    None
    """
    departures_by_month = df.groupby(["Year", "Month"])["Total"].sum().reset_index()
    departures_by_month = departures_by_month.pivot(index="Year", columns="Month", values="Total")
    
    departures_by_season = pd.DataFrame()
    departures_by_season["Winter"] = departures_by_month[12] + departures_by_month[1] + departures_by_month[2]
    departures_by_season["Spring"] = departures_by_month[3] + departures_by_month[4] + departures_by_month[5]
    departures_by_season["Summer"] = departures_by_month[6] + departures_by_month[7] + departures_by_month[8]
    departures_by_season["Autumn"] = departures_by_month[9] + departures_by_month[10] + departures_by_month[11]
    
    season_colors = ["#75aad8", "#ff89ff", "#ffd900", "#ff7400"]
    departures_by_season_sum = departures_by_season.sum(axis=0)
    
    fig, ax = plt.subplots(1, 2, figsize=(30, 6))
    departures_by_season.plot(kind="bar", stacked=False, color=season_colors, ax=ax[0], title="Flights per Season by Year", xlabel="Year", ylabel="Number of Flights")
    departures_by_season_sum.plot(kind="bar", color=season_colors, ax=ax[1], title="Total Flights per Season", xlabel="Season", ylabel="Number of Flights (in tens of millions)")
    
    plt.show()
```

```python
# Visualize flights per season
try:
    visualize_flights_per_season(df)
except Exception as e:
    print("An error occurred while visualizing flights per season:", str(e))
```

```python
def visualize_flights_by_carrier_group(df):
    """
    Visualize total flights by carrier group across years using a bar plot.

    Parameters:
    df (pd.DataFrame): DataFrame containing flight data.

    Returns:
    None
    """
    df_carriergroup = df.groupby(['Year', 'carriergroup'])['Total'].sum().reset_index()
    
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_carriergroup, x='Year', y='Total', hue='carriergroup')
    
    plt.title('Total by Carrier Group - Passengers Per Year')
    plt.xlabel('Year')
    plt.ylabel('Total')
    plt.legend(title='Carrier Group', loc='upper left')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Visualize total flights by carrier group
try:
    visualize_flights_by_carrier_group(df)
except Exception as e:
    print("An error occurred while visualizing total flights by carrier group:", str(e))
```

```python
def visualize_flights_per_year(df):
    """
    Visualize total flights per year using line plots.

    Parameters:
    df (pd.DataFrame): DataFrame containing flight data.

    Returns:
    None
    """
    df_scheduled = df.groupby('Year')['Scheduled'].sum().reset_index()
    df_chartered = df.groupby('Year')['Charter'].sum().reset_index()
    
    sns.set(style='whitegrid')
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Year', y='Scheduled', data=df_scheduled, label='Scheduled Flights', color='deepskyblue')
    sns.lineplot(x='Year', y='Charter', data=df_chartered, label='Chartered Flights', color='orange')
    
    plt.title('Total Flights per Year')
    plt.xlabel('Year')
    plt.ylabel('Total Flights (in Million)')
    plt.legend()
    plt.show()
```

```python
# Visualize total flights per year
try:
    visualize_flights_per_year(df)
except Exception as e:
    print("An error occurred while visualizing total flights per year:", str(e))
```
